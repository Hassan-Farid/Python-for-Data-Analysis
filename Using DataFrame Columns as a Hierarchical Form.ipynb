```python
#Sometimes we have a dataframe consisting of column values only and we need to convert it into a hierarchical form of dataframe
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
frame_obj = DataFrame({'a':np.arange(7), 'b':np.arange(7,0,-1), 'c':['one','one','one','two','two','two','two'], 'd':[0,1,2,0,1,2,3]})
frame_obj
```

```python
#Suppose we want to ease our operations so we convert the columns c and d into index labels for hierarchical dataframe
hr_frame_obj = frame_obj.set_index(['c','d']) #Takes a list of column values which are to be converted into index labels
hr_frame_obj
```

```python
#By default, the columns get removed and become the index labels
#To avoid this, we can use the drop argument in the set_index method
frame_obj.set_index(['c','d'], drop=False)
```

```python
#The inverse operation of the set_index is the reset_index method which converts the index labels of a hierarchical dataframe to a simple dataframe
hr_frame_obj.reset_index() #The reseted values are inserted from the left side
```
