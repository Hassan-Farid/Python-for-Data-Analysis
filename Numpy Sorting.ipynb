```python
#Like normal lists and arrays, we can sort the numpy arrays

import numpy as np
from numpy.random import randn

data = randn(8)
data
```

```python
#To sort elements we use: sort method
data.sort()
data
```

```python
#Additional argument can be given to indicate axis of sorting
data2 = randn(4,3,2)
data2
```

```python
data2.sort(2)
data2
```

```python
#Using sort method we can compute quantile of an array as
large_arr = randn(1000)
large_arr
```

```python
large_arr.sort()
large_arr
```

```python
large_arr[int(0.05 * len(large_arr))] #5% quantile
```
