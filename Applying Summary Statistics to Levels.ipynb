```python
#The summary statistics methods that were discusses previously can be applied to a certain levels
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
frame_obj = DataFrame(np.random.randn(12).reshape(4,3), index=[list('aabb'),[1,2,1,2]], columns=[['data1','data2','data2'],['first','first','second']])
frame_obj.index.names = ['key1','key2'] #Names the index label columns
frame_obj.columns.names = ['Datatype:', 'Dataset:'] #Names the value columns
frame_obj
```

```python
frame_obj.sum(level = 'key1') #Applies summation method onto the values of key1
```

```python
frame_obj.mean(level = 'key2') #Applies mean method onto the values in key2
```

```python
frame_obj.var(axis = 1, level = 'Dataset:') #Applies variance method onto the values in Dataset
```

```python
frame_obj.std(axis = 1, level='Datatype:') #Applies standard deviation method onto the values in Datatype
```

```python
frame_obj.describe()
```
