```python
#In case of hierarchical objects, we would sometimes have to rearrange and sort the data
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
frame_obj = DataFrame(np.random.randn(12).reshape(4,3), index=[list('aabb'),[1,2,1,2]], columns=[['data1','data2','data2'],['first','first','second']])
frame_obj.index.names = ['key1','key2'] #Names the index label columns
frame_obj.columns.names = ['Datatype:', 'Dataset:'] #Names the value columns
frame_obj
```

```python
#Suppose we want to interchange the index values, i.e. we want key2 to represent the 1st index label and key1 to represent the 2nd index label
frame_obj.swaplevel('key1','key2') #Swaps the two index labels without altering the values
```

```python
#In case we want to sort, the sort_index method allows to sort a single level lexicographically
frame_obj.sort_index()
```

```python
frame_obj.swaplevel(0,1).sort_index(1).sort_values(by = 'key2')
```
