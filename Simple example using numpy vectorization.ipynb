```python
import numpy as np

points = np.arange(-5, 5, 0.01) #1000 equally spaced points from -5 to 4
xs, ys = np.meshgrid(points, points) #Creates two 2D arrays corresponding to all pairs of (x,y) in the given 1D array arguments
```

```python
ys
```

```python
import matplotlib.pyplot as plt

z = np.sqrt(xs**2 + ys**2) #Initializing the value to be displayed
plt.imshow(z, cmap = plt.cm.gray) #Displays the visualization of the given function
plt.colorbar() #Displays the colorbar
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values") #Displays the title of the visualization image
```
